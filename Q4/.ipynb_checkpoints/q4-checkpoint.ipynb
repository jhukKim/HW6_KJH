```python
import csv
import matplotlib.pyplot as plt
import numpy as np


def main():
    plt.rc('font',family='Malgun Gothic')
    plt.rcParams['axes.unicode_minus']=False

    dic_1={}#승차
    dic_2={}#하차
    dic_3={}#승차+하차
    f=open('q4.csv')
    data=csv.reader(f)
    next(data)
    next(data)
    
    for row in data:
        num1=int(row[-4])+int(row[-2])
        num2=int(row[-3])+int(row[-1])
        num3=int(row[-4])+int(row[-2])+int(row[-3])+int(row[-1])
        dic_1[row[3]]=num1
        dic_2[row[3]]=num2
        dic_3[row[3]]=num3
        
    arr1=[] #30개 역이름 저장 리스트
    arr2=[] #30개 역 이용자수 저장 리스트
    
    
    lst1=list(dic_1)
    lst2=list(dic_1.values())
    
    for i in range(30):
        inum=lst2.index(max(lst2))
        arr1.append(lst1[inum])
        arr2.append(lst2[inum])
        
        lst2[inum]=-9999999
    plt.title('2023 3월 오전7~9시 출근시간 최대 승차역')
    plt.xlabel('역이름')
    plt.ylabel('(명)')
    plt.bar(arr1,arr2)
    plt.show()
    
    
    ###########################################
    arr1=[] #30개 역이름 저장 리스트
    arr2=[] #30개 역 이용자수 저장 리스트
    
    
    lst1=list(dic_2)
    lst2=list(dic_2.values())
    
    for i in range(30):
        inum=lst2.index(max(lst2))
        arr1.append(lst1[inum])
        arr2.append(lst2[inum])
        
        lst2[inum]=-9999999
    plt.title('2023 3월 오전7~9시 출근시간 최대 하차역')
    plt.xlabel('역이름')
    plt.ylabel('(명)')
    plt.bar(arr1,arr2)
    plt.show()
    
    #############################################
    arr1=[] #30개 역이름 저장 리스트
    arr2=[] #30개 역 이용자수 저장 리스트
    
    
    lst1=list(dic_3)
    lst2=list(dic_3.values())
    
    for i in range(30):
        inum=lst2.index(max(lst2))
        arr1.append(lst1[inum])
        arr2.append(lst2[inum])
        
        lst2[inum]=-9999999
    plt.title('2023 3월 오전7~9시 출근시간 최대 승하차역')
    plt.xlabel('역이름')
    plt.ylabel('(명)')
    plt.bar(arr1,arr2)
    plt.show()
        
        
    
    
    
    
if __name__=="__main__":
    main()
```
