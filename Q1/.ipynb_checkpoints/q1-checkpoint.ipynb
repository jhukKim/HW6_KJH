```python
import csv
import matplotlib.pyplot as plt

d_all={}
d_seoul={}
d_daejeon={}
d_pusan={}
d_jeju={}

for i in range (12):
    d_all[i+1]=0
    d_seoul[i+1]=0
    d_daejeon[i+1]=0
    d_pusan[i+1]=0
    d_jeju[i+1]=0
 
f=open('q1_all.csv')
data=csv.reader(f)

for i in range(7):
    next(data)
    
header=next(data)

num=0

for row in data:
    num+=1
    if len(row[2])!=0:
        d_all[num]=float(row[2])
f.close()


f=open('q1_seoul.csv')
data=csv.reader(f)

for i in range(7):
    next(data)
    
header=next(data)

num=0

for row in data:
    num+=1
    if len(row[2])!=0:
        d_seoul[num]=float(row[2])
f.close()

f=open('q1_daejeon.csv')
data=csv.reader(f)

for i in range(7):
    next(data)
    
header=next(data)

num=0

for row in data:
    num+=1
    if len(row[2])!=0:
        d_daejeon[num]=float(row[2])
f.close()

f=open('q1_pusan.csv')
data=csv.reader(f)

for i in range(7):
    next(data)
    
header=next(data)

num=0

for row in data:
    num+=1
    if len(row[2])!=0:
        d_pusan[num]=float(row[2])
f.close()

f=open('q1_jeju.csv')
data=csv.reader(f)

for i in range(7):
    next(data)
    
header=next(data)

num=0

for row in data:
    num+=1
    if len(row[2])!=0:
        d_jeju[num]=float(row[2])
f.close()

plt.title('Temp')

xdata=d_seoul.keys()


all_val=d_all.values()
seoul_val=d_seoul.values()
daejeon_val=d_daejeon.values()
pusan_val=d_pusan.values()
jeju_val=d_jeju.values()
plt.plot(xdata,all_val)
plt.plot(seoul_val)
plt.plot(daejeon_val)
plt.plot(pusan_val)
plt.plot(jeju_val)
plt.show()
```
