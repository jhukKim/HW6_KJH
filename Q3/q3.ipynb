```python
import csv
import matplotlib.pyplot as plt

def main():
    year=[2016,2017,2018,2019,2020,2021,2022]
    m=[]
    fm=[]
    mnum=2
    fnum=3
    f=open('q3.csv')
    data=csv.reader(f)
    next(data)
    next(data)
    for row in data:
        for i in range(7):
            m.append(int(row[mnum]))
            fm.append(int(row[fnum]))
            mnum+=4
            fnum+=4
            
    plt.title('2016~2022 Jeju Population')
    plt.xlabel('year')
    plt.ylabel('number of people')
    plt.plot(year,m,label='male')
    
    plt.plot(year,fm,label='female')
    
    plt.legend()
    plt.show()
        
        


if __name__=="__main__":
    main()
```
