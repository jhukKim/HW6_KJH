```python
import csv
import matplotlib.pyplot as plt

def main():

    num=[
        [],
        [],
        [],
        []
    ]

    f=open('q2.csv')
    data=csv.reader(f)

    a=0
    for row in data:
        for i in range(6):
            row[i]=int(row[i])
            for j in range(row[i]):
                num[a].append(i+1)
        plt.title('number of dice rolled: {}'.format(10**(a+2)))
        plt.xlabel('number')
        plt.ylabel('frequency')
        plt.hist(num[a],bins=100)
        plt.show()
        a+=1

    f.close()

if __name__=='__main__':
    main()
```
